PART B) Repeat Part A but use a normalized version of the data. Recall that one way to normalize the data is by subtracting the mean from the individual predictors and dividing by the standard deviation.

How does the mean of the mean squared errors compare to that from Step A?

0. Install and import necessarry libraries

```python
#!pip install numpy==1.21.4
#!pip install pandas==1.3.4
#pip install keras==2.1.6

import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

1) Load CSV Data "concrete_data.csv" and split data:
In Part B, the data will be normalised




```python

concreteData = pd.read_csv("concrete_data.csv") # Read the CSV Data

# Declare the predictors and target values in the table:
columnsAll = concreteData.columns # all the columns of the data

predictors = concreteData[columnsAll[columnsAll !='Strength']] # predictors are the columns except "Strength"
target = concreteData['Strength']

# Normalise the data:
predictorsNormalised = (predictors-predictors.mean()) / predictors.std()
targetNormalised     = (target-target.mean()) / target.std()

# Split the data as training data and test data:
X_train, X_test, y_train, y_test = train_test_split(predictorsNormalised,targetNormalised,test_size=0.3, random_state=42)


```

2) Train the model after creating it: But this time, 100 epochs

```python
# Create Model
nColumns = predictorsNormalised.shape[1] # number of columns of predictors

def regNN_model():
    model = Sequential() # Sequential Model
    model.add(Dense(10, activation='relu',input_shape=(nColumns,))) # 1. Hidden Layer with 10 Neurons. Input Layer should have so many neurons as columns of predictors.
    model.add(Dense(1)) # Output Layer which will give the predicted "Strength"

    model.compile(optimizer="adam", loss ="mean_squared_error") #Compile Model
    return model

model = regNN_model(); # Call the function to create the model
# Train the model
model.fit(X_train, y_train, epochs=100) # Make fitting
```

3. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. 

```python
predictedStrength = model.predict(X_test) #Predict the Strength with the test predictors
MSE=mean_squared_error(y_test,predictedStrength) # Calculate MSE with the methode
print("Mean Squared Error is: %f" %MSE) # Give the MSE
```

4) Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.

```python
MSE_array = np.zeros(50) # Create zero array
for i in range(50):
    X_train, X_test, y_train, y_test = train_test_split(predictorsNormalised,targetNormalised,test_size=0.3, random_state=42)
    model.fit(X_train, y_train, epochs=50) # Make fitting
    predictedStrength = model.predict(X_test)
    MSE_array[i]=(mean_squared_error(y_test,predictedStrength)) # Store the MSEs in the array
```

5. Report the mean and the standard deviation of the mean squared errors.

```python
from texttable import Texttable
standardDeviation_MSEArray = MSE_array.std()
mean_MSEArray = MSE_array.mean()

print("Mean of the Mean Square Error Array: %f" %mean_MSEArray)
print("Standard Deviation of the Mean Square Error is: %f" %standardDeviation_MSEArray)

t = Texttable()
t.add_rows([['Mean of the List of MSEs', 'Standard Deviatiion of the List of MSEs'], [mean_MSEArray,standardDeviation_MSEArray]])
print(t.draw())

```

RESULT: We see that, the change in mean square error is very small with compared to PART B
